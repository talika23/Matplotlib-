```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2 * np.pi, 100)

# Calculate the sine of each value
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add labels and a title
plt.title('Sine Wave')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')

# Show the plot
plt.grid(True)
plt.show()
```
