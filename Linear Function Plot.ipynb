```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data: y = 2x + 1
x = np.linspace(-10, 10, 100)
y = 2 * x + 1

# Plot the data
plt.plot(x, y, label='y = 2x + 1')

# Add labels and title
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
