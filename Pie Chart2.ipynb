```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([10,20,30,40])
mylabels = ["Sweety", "Sneha", "Arti", "Yogita"]
myexplode = [0.2, 0, 0, 0]
mycolors = ['red','black','yellow','green']
plt.pie(y, labels = mylabels, explode = myexplode, shadow = 'true', colors = mycolors)

plt.legend()
plt.show() 
```
